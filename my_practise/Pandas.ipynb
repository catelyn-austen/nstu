# Pandas
* нужна для интуитивно понятной обработки данных, особенно из файлов типа ecxel, json
* pd.DataFrame([])
* dataframe - совокупность массивов, которые называются series (посл-ть)
* df.columns = ['name', 'age'] - даем название столбцам
* df.iloc[row, column] - обращение к элементу по индексу
* после именования столбца, уже нельзя обращаться по его индексу, но зато можно брать срезом (df[0:2])

* df[['name', 'age']] - выбор столбцов по названиям
* df.index = df['name'] - даем название строкам
* df.loc['row', 'column'] - обращение к элементу по названию
* можно также делать текстовые срезы, только __конец среза включается__
* df.loc['sam':'lisa', ['age', 'height']] - срез по названию строк + конкретные столбцы
* df.column.hist(bins = 100) - строим график из 100 значений

#### Функции lambda
* f = __lambda__ x: x.max() - x.min() - пример

```python
import numpy as np
import pandas as pd

f = lambda x: x ** 2
l = [1, 2, 3]
res = list(map(f, l))
print(res)

f1 = lambda x: x.min() - x.max()
l_arr = np.array(l)
res = f1(l_arr)
print(res)
```

#### Метод apply()
* метод применяется для работы с большим количеством столбцов
* df['name'] = df['name1'].apply(np.log, axis=1) - формирование нового столбца на основе старого так, что от значений старого берется логарифм. axis=1 указывает на то, что операция проводится построчно
* df['name_n'] = df.loc[:, 'name':'namen'].apply(np.mean, axis=1) - создание нового столбца на основе нескольких
* вместо np.func может быть любая функция

#### Группирование и агрегирование
* groupby группирует данные по столбцу
* df.groupby('class')
* groupby не вернет список, поэтому нужно преобразовать результат в list: list(df.groupby('class'))


* агрегирование помогает увидеть особенности каждой группы
* df.groupby('class').agg('sum') - сгруппировали по классам и просуммировали остальные столбцы
* если хочется посмотреть еще функции, характеризующие группы (т.е. больше одной), то их нужно указывать в виде списка: agg(['sum', 'mean', 'min'])

#### Сортировка и упорядочивание
* df.sort_values(['', '', ...], ascending=True/False, inplace=True/False) - метод, сортирующий по конкретным столбцам
* ascending - по возрастанию, если True
* inplace - сохранять ли изменения в дф
* первый столбец в приоритете, потом второй, потом третий, и т.д.

#### Работа с NaN
* isnull() - показывает, где значения не заполнены
* isnull().sum() - покажет, сколько NaN в каждом столбце
* fillna(что) - заменяет NaN на что-то, указанное в скобках
* если данных очень много, и от потери нескольких строк ничего не изменится, то можно воспользоваться удалением этих нановых строк:

dropna()

#### Иерархическое индексирование
* set_index(['', '', ''...]) - метода, который применяется для изменения дефолтного индекса на тот, который мы хотим сделать. его можно создать, объединив несколько строк
* sort_index() - метод, сортирующий строки по составным индексам


```python
kaka = pd.DataFrame([['Catelyn', 'Austen', 20],
                     ['Yuru', 'Azumi', 20],
                     ['Alice', 'Azumi', 0],
                     ['Richard', 'Austen', 0],
                     ['Alice', 'Austen', 0]], columns = ['name', 'lastname', 'age'])
kaka.set_index(['name', 'lastname'], inplace=True)
```

```python
# сортировка по индексам
kaka.sort_index()
```

```python
# обращение к строкам через составные индексы
print(kaka.loc['Alice'], '\n')
print(kaka.loc['Alice', 'Azumi'])
```
