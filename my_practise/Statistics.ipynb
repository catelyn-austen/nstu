#### Плотность распределения
Это функция, характеризующая сравнительную вероятность попадания случайной величины на участок $$ x + \delta x $$
График: __гистограмма__ или __кривая распределения__
Пригодна как для дискр, так и для непрер величин

#### Центральная предельная теорема
*Сумма независимых одинаково распределенных случайных величин имеет распределение близкое к нормальному* (крч говоря оно встречается часто)

#### Функция распределения
Характерна только для непрерывных случайных величин и характеризует вероятность попадания элемента х в интервал



#### Точечные оценки СВ
* Мат ожидание (среднее)
* Дисперсия (мера разброса) - показатель рассеяния значений вокруг среднего СВ



* Среднеквардратичное отклонение (сигма)



* Мода (Мо(х)) - самое часто встречающееся значение СВ
* Квантиль ($ x_\alpha $) - оценки, отвечающие на вопрос: какое значение СВ не привысит заданной вероятности (25, 50, 75 процентов, например)
* Квартиль - квантили с вероятностями 25, 50, 75 процентов, они разделяют все исходы на 4 равных отрезках. 50% - медиана
* Интерквартильный размах (IQR) - аналог дисперсии для оценки разброса СВ. Разница м/у 0,75 и 0,25 квантилем: $ x_75 - x_25 $



##### Остроконечность
* Эксцесс - это мера островершинности распределения. Он показывает, насколько «тяжелые хвосты» у распределения по сравнению с нормальным распределением.  

Интерпретация:

 Эксцесс < 0: Распределение имеет более высокие пики и более тяжелые хвосты, чем нормальное распределение. Это означает, что у распределения больше крайних значений, чем у нормального распределения.

 Эксцесс > 0: Распределение имеет более низкие пики и более легкие хвосты, чем нормальное распределение. Это означает, что у распределения меньше крайних значений, чем у нормального распределения.

 Эксцесс = 0: Распределение является нормальным (гауссовым) распределением.




##### Скос
* Асимметрия — это мера  несимметричности распределения. Она показывает, насколько «скошен» график распределения относительно своего центра. 

Интерпретация:

 Асимметрия > 0: Распределение скошено вправо (в сторону больших значений). Хвост распределения длиннее в правой части, а пик распределения смещен влево.

 Асимметрия < 0: Распределение скошено влево (в сторону меньших значений). Хвост распределения длиннее в левой части, а пик распределения смещен вправо.

 Асимметрия = 0: Распределение симметрично относительно своего центра.

#### Правило 3х сигм
Для нормального распределения данных вероятность выхода за коридор $ m(x) +- 3 \sigma $ практически = 0. Все значения, выпрыгнувшие за сигмы, являются аномальными

# Как считать порядковый номер элемента при подсчете квантиля
Допустим, хотим найти $$ \alpha = 0.2 $$ квантиль в выборке n элементов. Для этого просто $$ \alpha * n $$ и получаем нужный номер. Если это число будет дробным, например, 6.6, то мы возьмем элементы с порядковым номером 6 и 7 и найдем среднее этих элементов.

#### Квантиль - в долях, перцентиль - в процентах, хотя смысл один. А вот квартиль - частный случай (как и медиана) квантиля, который делит выборку на 4 равные части. К квартилям относят Q1=0.25, Q2=0.5=медиана, Q3=0.75. Наиболее стабильная часть выборки лежит между Q1 и Q2.
